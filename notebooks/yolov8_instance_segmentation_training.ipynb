```python

from roboflow import Roboflow
rf = Roboflow(api_key="(your api key of roboflow)")
project = rf.workspace("pankajs-lab").project("potholedetection-instancesegmentation")
version = project.version(4)
dataset = version.download("yolov8")
```

```python
from ultralytics import YOLO
model=YOLO("yolov8n-seg.pt")
result = model.train(
    data="/content/potholedetection-instancesegmentation-4/data.yaml",
    epochs=80,
    imgsz=768,
    batch=16,
    cache=True,
    patience=50,
    lr0=0.001,
    optimizer="AdamW",
    augment=True
)
```

```python
from ultralytics import YOLO

model=YOLO("best.pt")
metrics = model.val(
    data="/content/potholedetection-instancesegmentation-4/data.yaml",
    split="val",
    plots=True,
    save_json=True
)
```

```python
metrics = model.val()

print(metrics.box.map)      # bbox mAP
print(metrics.seg.map)      # segmentation mAP
print(metrics.box.map50)
print(metrics.seg.map50)
```
